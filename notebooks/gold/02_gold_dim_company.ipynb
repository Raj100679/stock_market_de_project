```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from datetime import datetime
from pyspark.sql.window import Window
```

```python
silver_company_df = spark.read.table("raj.silver.company_info")
silver_company_df.printSchema()
```

```python
silver_company_df = silver_company_df \
    .withColumn("company_key", row_number().over(Window.orderBy("ticker"))) \
    .withColumn("start_date", current_date()) \
    .withColumn("end_date", lit(None).cast(DateType())) \
    .withColumn("is_current", lit(True)) \
    .drop("silver_timestamp", "is_valid")
display(silver_company_df)
```

```python
silver_company_df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("raj.gold.dim_company")
```
